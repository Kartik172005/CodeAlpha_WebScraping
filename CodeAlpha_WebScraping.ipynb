```python
# ==========================================================
# CodeAlpha Task 1: Web Scraping
# Project: Books to Scrape Web Scraper
# Author: Your Name
# ==========================================================

# Install required libraries
!pip -q install requests beautifulsoup4 pandas lxml

import requests
from bs4 import BeautifulSoup
import pandas as pd
from google.colab import files

BASE_URL = "https://books.toscrape.com/"

books_data = []

print("========================================")
print("   CodeAlpha Web Scraping Project")
print("========================================\n")

for page in range(1, 51):

    if page == 1:
        url = BASE_URL
    else:
        url = BASE_URL + f"catalogue/page-{page}.html"

    print(f"Scraping Page {page}/50")

    response = requests.get(url)
    response.encoding = "utf-8"

    if response.status_code != 200:
        print(f"Could not access Page {page}")
        continue

    soup = BeautifulSoup(response.text, "lxml")

    books = soup.find_all("article", class_="product_pod")

    for book in books:

        # Title
        title = book.h3.a["title"]

        # Price
        price = book.find("p", class_="price_color").text
        price = (
            price.replace("Â£", "")
                 .replace("£", "")
                 .replace("Â", "")
                 .strip()
        )

        # Rating
        rating = book.find("p", class_="star-rating")["class"][1]

        # Availability
        availability = book.find(
            "p",
            class_="instock availability"
        ).text.strip()

        # Product Link
        product_link = BASE_URL + "catalogue/" + book.h3.a["href"].replace("../", "")

        books_data.append({
            "Title": title,
            "Price (£)": price,
            "Rating": rating,
            "Availability": availability,
            "Product URL": product_link
        })

# Create DataFrame
df = pd.DataFrame(books_data)

print("\n========================================")
print("Web Scraping Completed Successfully!")
print("========================================")

print(f"\nTotal Books Collected: {len(df)}")

# Show first 10 records
display(df.head(10))

# Save CSV
csv_filename = "books_dataset.csv"
df.to_csv(csv_filename, index=False)

print(f"\nCSV File Saved: {csv_filename}")

# Download CSV
files.download(csv_filename)
```
